```python
import requests
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
url = "https://api.worldbank.org/v2/country/COL/indicator/FP.CPI.TOTL.ZG?format=json&per_page=100"
```

```python
respuesta = requests.get(url)
```

```python
if respuesta.status_code == 200:
    data = respuesta.json()
    registros = data[1]

    df_ipc = pd.DataFrame(registros)[["date","value"]]
    df_ipc.columns = ["A単o", "IPC (%)"]
    df_ipc = df_ipc.dropna().sort_values("A単o", ascending=False)
    print(df_ipc.head())
else:
    print("Error al consultar la API:", respuesta.status_code)
```

```python
df_ipc
```

```python
sns.set(style="whitegrid")
plt.figure(figsize=(12,6))
sns.lineplot(data=df_ipc, x="A単o", y="IPC (%)", marker="o")
plt.title("Evoluci坦n hist坦rica del IPC en Colombia")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
